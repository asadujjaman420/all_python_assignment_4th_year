Q1

```python
with open("input.txt") as f:
    lines = f.readlines()

print("Roll  inMark1  in_Mark2    Final  Grade  GradePoint")
for line in lines[:]:
    roll, m1, m2, m3 = line.split()
    m1, m2, m3 = float(m1), float(m2), float(m3)
    final = m3+(m1+m2)*.5
    if final >= 80:
        grade, point = "A+", 4.0
    elif final >= 75:
        grade, point = "A", 3.75
    elif final >= 70:
        grade, point = "A-", 3.5
    elif final >= 65:
        grade, point = "B+", 3.25
    elif final >= 60:
        grade, point = "B", 3.0
    elif final >= 55:
        grade, point = "B-", 2.75
    elif final >= 50:
        grade, point = "C+", 2.5
    elif final >= 45:
        grade, point = "C", 2.25
    elif final >= 40:
        grade, point = "D", 2.0
    else:
        grade, point = "F", 0.0

    print(f"{roll}  {m1}  {m2}  {m3}  {final:.2f}  {grade}  {point}")
```

```python
word = []
teachers = open('amth.txt')

for teacher in teachers:
    teacher = teacher.strip()
    words = teacher.split()
    for word_list in words:
        if word_list not in word:  # Avoid duplicates
            word.append(word_list)

print(word)

word.sort()
print(word)
```

Q4

```python

file = open('mbox-short.txt')

total = 0.0
count = 0


for line in file:
    if line.startswith("X-DSPAM-Confidence:"):
        value = float(line.split(":")[1].strip())
        total += value
        count += 1

if count > 0:
    average = total / count
else:
    average = 0.0


print(f"Average spam confidence: {average:.4f}")
```

```python
file=open('mbox-short.txt')
count=0
email_list=[]
for line in file:
    line=line.strip()
    if line.startswith('From '):
        count+=1
        words=line.split()
        if words[1] not in email_list:
            email_list.append(words[1])


print(count)
print(email_list)
```
