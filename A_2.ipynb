```python
#q5_b
import sympy as sp
x,y,z,lam=sp.symbols('x,y,z,lam',real=True)

f1=4*x**2+y**2+4*z**2-16
T=8*x**2+4*y*z-16*z+600
eq1=sp.Eq(sp.diff(T,x)-lam*sp.diff(f1,x),0)
eq2=sp.Eq(sp.diff(T,y)-lam*sp.diff(f1,y),0)
eq3=sp.Eq(sp.diff(T,z)-lam*sp.diff(f1,z),0)
eq4=sp.Eq(f1,0)
soln=sp.solve([eq1,eq2,eq3,eq4],[x,y,z,lam])

print(soln)
max_T=None
max_point=None
for s in soln:
    T_val=T.subs({x:s[0],y:s[1],z:s[2],lam:s[3]})
    if max_T is None or T_val>max_T:
        max_T=T_val
        max_point=s
print("Hottest point:", max_point)
print("Maximum Temperature:",float(max_T))
```

```python
##stock theorem verification
import sympy as sp
from sympy.vector import CoordSys3D,curl
from sympy import integrate
x,y,z,r,t=sp.symbols('x y z r t',real=True)
N=CoordSys3D('N')
F=2*N.z*N.i+3*N.x*N.j+5*N.y*N.k
curl_F=curl(F)
curl_F=curl_F.subs({N.x:r*sp.cos(t),N.y:r*sp.sin(t),N.z:4-r**2})
curl_F = sp.Matrix([
    curl_F.dot(N.i),
    curl_F.dot(N.j),
    curl_F.dot(N.k)
])
p=sp.Matrix([r*sp.cos(t),r*sp.sin(t),4-r**2])
p_r=sp.diff(p,r)
p_t=sp.diff(p,t)
f=(p_r.cross(p_t)).dot(curl_F)
I=integrate(f,(r,0,2),(t,0,2*sp.pi))
print(I)
#line integral
x=2*sp.cos(t)
y=2*sp.sin(t)
z=0
F=F.subs({N.x:x,N.y:y,N.z:z})
F=sp.Matrix([
    F.dot(N.i),
    F.dot(N.j),
    F.dot(N.k)
])
r=sp.Matrix([x,y,z])
r_prime=r.diff(t)
f=F.dot(r_prime)
I=integrate(f,(t,0,2*sp.pi))
print(I)
```

```python
#surface integral
import sympy as sp
from sympy.vector import CoordSys3D,curl

N=CoordSys3D('N')
x,y,z,r,t=sp.symbols('x y z r t',real=True)
F=2*N.z*N.i+3*N.x*N.j+5*N.y*N.k
curl_F=curl(F)
print(curl_F)

curl_F=curl_F.subs({N.x:r*sp.cos(t),N.y:r*sp.sin(t),N.z:4-r**2})
curl_F=sp.Matrix([
    curl_F.dot(N.i),
    curl_F.dot(N.j),
    curl_F.dot(N.k)
])
p=sp.Matrix([r*sp.cos(t),r*sp.sin(t),4-r**2])
p_r=sp.diff(p,r)
p_t=sp.diff(p,t)

normal_vector=p_r.cross(p_t)
integrand=curl_F.dot(normal_vector)
I=integrate(integrand,(r,0,2),(t,0,2*sp.pi))
print(I)
```
